# 1. Importing Necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 2. Data cleaning and Exploratory data analysis

```python
df = pd.read_csv('./bank.csv')
df.head()
```

## Check the unique values 

```python
df['Country Name'].unique()
```

## Understanding values
Figuring out the values that are present in the dataset. If there are null values we shall remove them. Its handy to check the data type of the data columns and values. 

```python
df = df.iloc[:, :-1]
df.info()

df.isna().count()
```

```python
df['Country Code'].nunique()
```

```python
df['Country Name'].nunique()
```

# 3. Visualisation 
Since the no. of region and countries is around 266. Visualising these no. of countries is impractical. For which we opt to visualise by region and by continents.

## Visualisation of South Asia

```python
south_asia_population = df[df["Country Name"] == 'South Asia (IDA & IBRD)']
years = south_asia_population.columns[5:]
years = years.astype(str).tolist()  
population_over_period = south_asia_population.iloc[:, 5:].squeeze()
country_name = south_asia_population.iloc[0]['Country Name']

plt.figure(figsize=(10, 2))
plt.bar(years, population_over_period, color='blue', alpha=0.7)
plt.title('Population of South Asia')
plt.xlabel('Years')
plt.ylabel('Population in Billions')
plt.grid(axis='y')
plt.xticks(years, rotation=90)
plt.show()
```
